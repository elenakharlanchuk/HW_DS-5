```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
#Распределение Пуассона
mu=1.4
#задаем распределение Пуассона
poisson_rv=sts.poisson(mu)

# сгенерируем 100 значений
poisson_rv.rvs(1000)
```

```python
#Построим график функции распределения Пуассона:
poisson_rv = sts.poisson(1)
poisson_rv2 = sts.poisson(2)
poisson_rv3 = sts.poisson(3)
poisson_rv4 = sts.poisson(4)
x = np.linspace(0,20,21)
cdf = poisson_rv.cdf(x) 
cdf2 = poisson_rv2.cdf(x) 
cdf3 = poisson_rv3.cdf(x) 
cdf4 = poisson_rv4.cdf(x) 
#plt.figure(figsize=(15,8))
plt.plot(x, cdf)
plt.plot(x, cdf2)
plt.plot(x, cdf3)
plt.plot(x, cdf4)

plt.legend(['mu=1',  'mu=2', 'mu=3', 'mu=4'])
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
##Построим график плотности распределения Пуассона:

poisson_rv = sts.poisson(1)
poisson_rv2 = sts.poisson(2)
poisson_rv3 = sts.poisson(3)
poisson_rv4 = sts.poisson(4)
x = np.linspace(0,20,21)
pmf = poisson_rv.pmf(x)
pmf2 = poisson_rv2.pmf(x)
pmf3 = poisson_rv3.pmf(x)
pmf4 = poisson_rv4.pmf(x)
#plt.figure(figsize=(15,8))
plt.plot(x, pmf, 'o')
plt.plot(x, pmf2,'o')
plt.plot(x, pmf3,'o')
plt.plot(x, pmf4,'o')

plt.legend(['mu=1',  'mu=2', 'mu=3', 'mu=4'])
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
#Экспоненциальное распределение (sts.expon)
```

```python
#задаем экспоненциальное распределение
#lambda = 1.2
expon_rv=sts.expon(1.2)

# сгенерируем 100 значений
poisson_rv.rvs(1000)
```

```python
#Построим график экспоненциальной функции распределения:
expon_rv = sts.expon(1)
expon_rv2 = sts.expon(2)
expon_rv3 = sts.expon(3)
expon_rv4 = sts.expon(4)
x = np.linspace(3,20,21)
cdf = expon_rv.cdf(x) 
cdf2 = expon_rv2.cdf(x) 
cdf3 = expon_rv3.cdf(x) 
cdf4 = expon_rv4.cdf(x) 
#plt.figure(figsize=(15,8))
plt.plot(x, cdf)
plt.plot(x, cdf2)
plt.plot(x, cdf3)
plt.plot(x, cdf4)

plt.legend(['lambda=1',  'lambda=2', 'lambda=3', 'lambda=4'])
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
##Построим график плотности экспоненциального распределения :

expon_rv = sts.expon(1)
expon_rv2 = sts.expon(2)
expon_rv3 = sts.expon(3)
expon_rv4 = sts.expon(4)
x = np.linspace(5,20,30)
pdf = expon_rv.pdf(x)
pdf2 = expon_rv2.pdf(x)
pdf3 = expon_rv3.pdf(x)
pdf4 = expon_rv4.pdf(x)
#plt.figure(figsize=(15,8))
plt.plot(x, pdf)
plt.plot(x, pdf2)
plt.plot(x, pdf3)
plt.plot(x, pdf4)

plt.legend(['lambda=1',  'lambda=2', 'lambda=3', 'lambda=4'])
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
#распределение Лапласа
```

```python
#задаем распределение Лапласа
#lambda = 1.2
laplace_rv=sts.laplace(1,0)

# сгенерируем 100 значений
laplace_rv.rvs(1000)
```

```python
#Построим график функции распределения Лапласа:
laplace_rv = sts.laplace(0.4)
laplace_rv2 = sts.laplace(0.8)
laplace_rv3 = sts.laplace(1)
laplace_rv4 = sts.laplace(2)
x = np.linspace(-10,20,10)
cdf = laplace_rv.cdf(x) 
cdf2 = laplace_rv2.cdf(x) 
cdf3 = laplace_rv3.cdf(x) 
cdf4 = laplace_rv4.cdf(x) 
#plt.figure(figsize=(15,8))
plt.plot(x, cdf)
plt.plot(x, cdf2)
plt.plot(x, cdf3)
plt.plot(x, cdf4)

plt.legend(['a=0.4',  'a=0.8', 'a=1', 'a=2'])
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```
